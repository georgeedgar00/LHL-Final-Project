```python
import pandas as pd 
import numpy as np 
import seaborn as sns  
import matplotlib.pyplot as plt 
#import plotly
```

```python
df = pd.read_csv("data_cleaned_2021.csv")
```

```python
df
```

# Salary parsing

```python
# Show if 'per hour' is included in 'Salary Estimate'
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if'per hour'in x.lower() else 0)
```

```python
# Show if 'employer provided salary' is included in 'Salary Estimate'
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if'employer provided salary'in x.lower() else 0)
```

```python
df
```

```python
# New column called 'salary' and removing '(Glassdoor est.)' 
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
# Replace '$' and 'K' in with empty string
remove_Kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
remove_Kd
```

```python
# Replace 'per hour' and 'employer provided salary:' with empty string
remove_hr = remove_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
remove_hr
```

```python
# Add min_salary, max_salary, avg_salary to table and change to int type
df['min_salary'] = remove_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = remove_hr.apply(lambda x: int(x.split('-')[1]))
df['avg_salary'] = (df['min_salary'] + df['max_salary'])/2
df
```

# Company name text only

```python
# Add company name and remove everything after name
df['company_text'] = df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-5], axis = 1)
df
```

# State field

```python
# Add state of job
df['job_state'] = df['Location'].apply(lambda x: x.split(', ')[1])
df
```

```python
# Count job listings by state
df.job_state.value_counts()
```

```python
# Check if job is in headquarters
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
df
```

# Age of company

```python
df['age'] = df.Founded.apply(lambda x: x if x < 1 else 2023 - x)
df
```

# Parsing of job description

Seeing if python, R, spark, aws and excel are in the job descriptions

```python
#Python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df.python.value_counts()
```

```python
# R
df['R'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
df.R.value_counts()
```

```python
# Spark
df['spark_yn'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df.spark_yn.value_counts()
```

```python
# AWS
df['AWS'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df.AWS.value_counts()
```

```python
# Excel
df['excel_yn'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df.excel_yn.value_counts()
```
